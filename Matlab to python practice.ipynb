```python
import numpy as np
from numpy import array
import scipy.linalg

# example array a 
a = np.array([[[1],[2],[3]],[[4],[5],[6]]]) # multi-dimentional array literal
b= np.matrix("[1,2,3 ; 4,5,6]") # matrix literal
c1 = array([1,2,3])
c2 = array([4,5,6])
d = np.random.randint(0,100,[3,5]) # random 3*5 matrix
A = np.random.randint(1,10,[3,3])
x = [1,2,3]
B = A@x
```

```python
print(np.ndim(a)) # num of dims
print(a.ndim)
print()
print(np.size(a)) # num of elements
print(a.size)
print()
print(np.shape(b)) # size of matrix
print(b.shape)
print()
print(a.shape[3-1]) # num of elemets of n-th dim of array
print()
c=np.block([[c1],[c2]])
print(c) # construct matrix from blocks
print()
print(c1[-1]) # last elemet of 1*n matrix c1
print()
print(c[1,2]) # access elemet at (row 2, col 3)
print()
print(c[0]) # entire first row
print()
print(c[0:2]) # entire first two row
print()
print(c[-2:]) # entire last two row
print()
print(d)
print(d[np.ix_([0,2],[0,1,3])]) # elements across row 1,3 and col 1,2,4
print(d[:,0:4:2]) # every other cols from col1 to col 5
print(d[:,::-1]) # reverse col order
print(d[np.r_[:len(d),1]]) # copy row two and append to the end
```

```python
print(c.T) # Transpose
print(c.conj().T) # conjugate transpose
print()
print(c@c1) # matrix multiply
print(c1*c) # element-wise multiply
print(c*6/c1) # element-wise divide
print(c**2) # element-wise exponentiation
print()
print((c>3)) # logic greater than
print(np.nonzero(c>3)) # index of elements greater than 3
print(c[:,np.nonzero(c1>2)[0]]) # extract the col of c where c1>2
print(c[:,c1.T>2]) # extrace col of c there c1.T>2
e = c.copy() # numpy assigns by reference
e[e<2] = 0 # zero out elements smaller than 2
print(e)
print()
f = c[1,:].copy() # numpy slices are by reference
print(f)
print()
g=c.flatten() # turn array into vector (forces a copy)
print(g)
```

```python
print(np.arange(1.,11.)) # create an increasing vector
print(np.r_[1. : 11])
print(np.r_[1.:10.:10j])
print(np.arange(10.))
print(np.r_[:10.])
print(np.r_[:9:10j])
print(np.arange(1.,11.)[:,np.newaxis]) # create an increase col
print()
print(np.zeros((3,4))) # create 3*4 zero array
print(np.zeros((3,4,5))) # 3*4*5 zero array
print()
print(np.ones((3,4))) # create 3*4 two dim array full of 64 bit floating point ones
print()
print(np.eye(3)) # 3*3 identity matrix
print(np.diag(c1)) # diagonal matrix of c1
print(np.diag(c1,1)) # diagonal matrix, diagonal above the main diagonal
print()
print(np.linspace(1,3,4)) # 4 equally spaced samples between 1 and 3
```

```python
print(np.mgrid[0:9.,0:6.]) # 2D arrays: one of x values, the other of y values
print(np.meshgrid(np.r_[0:9.],np.r_[0:6.]))
print(np.ogrid[0:9.,0:6]) # best way to eval functions on a grid
print(np.ix_(np.r_[0:9.],np.r_[0:6.]))
print()
print(np.meshgrid([1,2,4],[2,4,5]))
print(np.ix_([1,2,4],[2,4,5]))
```

```python
print(np.tile(c,(2,3))) # create a 2*3 copy of c
print()
print(np.concatenate((c,e),1)) # concatenate c and e
print(np.hstack((c,e)))
print(np.column_stack((c,e)))
print(np.c_[c,e])
print(np.concatenate((c,e))) # concatenate rows of c and e
print(np.vstack((c,e)))
print(np.r_[c,e])
```

```python
print(c.max()) # return the maximum element of c
print(c.max(0)) # return the maximum element of each col of matrix c
print(c.max(1)) # return the maximum element of each row of matrix c
print(np.maximum(c,e)) 
# compare c and e element-wise, and return the maximum value from each pair
print()
print(np.sqrt(c1@c1)) # L2 norm
print(np.linalg.norm(c1))
print()
print(np.logical_and(c,e)) # element-wise AND operator
print(c&e) # bitwise AND operator
print(c|e) # bitwise OR operator
```

```python
print(np.linalg.inv(np.vstack((c,c1-0.5)))) # inverse of square matrix
print(np.linalg.pinv(c)) # pseudo-inverse of matrix c
print()
print(np.linalg.matrix_rank(c)) # matrix rank of array c
print()
print(np.linalg.solve(A,B)) # solve for x in Ax=B (when A is square)
print(np.linalg.lstsq(A,B,rcond=None)) # solve for x in Ax=B (when A is not square) 
# multilinear regression
print(np.linalg.solve(A.T,B.T)) # solve for x in xA=B
print()
U,S,Vh = np.linalg.svd(A) # SVD
V = Vh.T
print ("U =",U,"\n","S =",S,"\n", "V=", V)
print()
print(np.linalg.cholesky([[2,-1],[-1,2]]).T) # cholesky facterization
print()
D,V = np.linalg.eig(A) # eigen decomposition
print("D =",D,"\n V =",V)
print()
Q,R = scipy.linalg.qr(A) # QR decomposition
print("Q =",Q,"\n R =",R)
print()
L1,U1,P = scipy.linalg.lu(A) # LU decomposition
print("L =",L1,"\n U =",U1,"\n P =",P)
print()
import scipy.sparse.linalg
print(scipy.sparse.linalg.cg(A,B)) # conjugate gradients solver
```

```python
print(np.fft.fft(c1)) # Fourier trans
print(np.fft.ifft(np.fft.fft(c1))) # inverse Fourier trans
print()
print(np.sort(A)) # sort a matrix
A.sort()
print(A)
print()
I = np.argsort(A[:,2]) # sort one row at a time
b=A[2,:]
print(b)
print()
import scipy.signal
scipy.signal.resample(A.flatten(),int(len(A.flatten())/2)) # downsample
print()
print(np.unique(A)) # find unique elements in A
print()
print(a.shape)
print(np.squeeze(a,axis=2)) # squeeze the dim with dim = 1 
```
